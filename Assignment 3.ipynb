```python
df = Data.copy()
```

```python
New = df.drop(["Id", "CITY"], axis=1)
New.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categ = ['Married/Single','House_Ownership','Car_Ownership','Profession','STATE']
le = LabelEncoder()
New[categ] = New[categ].apply(le.fit_transform)
```

```python
New.head()
```

```python
X = New.drop('Risk_Flag',axis=1)
y = New['Risk_Flag']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.head()
```

```python
y.head()
```

```python
#Split your dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = dtree.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
import six
import sys
sys.modules['sklearn.externals.six'] = six
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
import pydot 
```

```python
features = list(New.columns[1:])
features
```

```python
dot_data = StringIO()  
export_graphviz(dtree, out_file=dot_data,feature_names=features,filled=True,rounded=True)

graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png()) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(y_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test,rfc_pred))
```
